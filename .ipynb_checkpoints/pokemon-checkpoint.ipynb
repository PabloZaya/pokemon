```python
#para imprimir un df demasiado grande para ver en un intento
#for poke in pd.read_csv('Pokemon.csv', chunksize=5):
 #   print('CHUNK POKE')
  #  print(poke)
#para procesar los resultados y luego sumarlos a un nuevo df:
new_poke=pd.DataFrame(columns=poke.columns)
for poke in pd.read_csv('Pokemon.csv', chunksize=5):
    results=poke.groupby(['Type 1']).count()
    
    new_poke=pd.concat([new_poke, results])
```
